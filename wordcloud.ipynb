```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline

from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
ws = WS("./data")
pos = POS("./data")
ner = NER("./data")
```

# NCCU

```python
nccu = pd.read_csv('nccu_title.xlsx')
```

```python
def clean_text_and_cut(text):
    text = re.sub(r'[^\w\s]','',text) #remove all punctuations
    text = text.replace('\n', '').replace('\r', '').replace('\t', '').replace(' ', '').replace('[', '').replace(']', '')
    text = ws([text])
    return text
```

```python
all_text = []
for i in range(len(nccu)):
    title = clean_text_and_cut(nccu['title'][i])[0]
    all_text = all_text + title

pos_results = pos(all_text)
```

```python
all_text
```

```python
pos_results
```

```python

ner_results = ner(all_text, pos_results)
```

```python
new_all_text = []
for i in range(len(pos_results)):
    if ('A' in pos_results[i]) or ('Na' in pos_results[i]) or ('Nb' in pos_results[i]):
        new_all_text.append(all_text[i])
```

```python
stopw = ['人','時間','如果','知道','明天','成績','老師','課程','問題','政大','學期','申請','課','作業','上課',\
        '報告','莊','結果','畢業']
content = ' '.join(new_all_text)
wc = WordCloud(
            font_path = 'SourceHanSansTW-Regular.otf',
            width=2000, height=1200,
            stopwords = stopw)
cloud = wc.generate(content)
wc.to_file("nccu_hot_words.jpg")

plt.imshow(cloud)
plt.axis("off")
plt.show()
```
